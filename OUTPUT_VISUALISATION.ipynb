```python
import plotly.graph_objects as go
from collections import defaultdict

def read_input(file_path):
    with open(file_path, 'r') as file:
        # Read Stellar Crystals
        n = int(file.readline().strip())
        stellar_crystals = []
        for _ in range(n):
            x, y, c = map(int, file.readline().strip().split())
            stellar_crystals.append((x, y, c))

        # Read Void Mines
        m = int(file.readline().strip())
        void_mines = []
        for _ in range(m):
            x, y, m_val = map(int, file.readline().strip().split())
            void_mines.append((x, y, m_val))

    return stellar_crystals, void_mines

def read_output(file_path):
    with open(file_path, 'r') as file:
        cost = int(file.readline().strip())
        v, e = map(int, file.readline().strip().split(','))
        edges = []
        for _ in range(e):
            line = file.readline().strip()
            # Remove parentheses and split by commas
            coords = line.replace('(', '').replace(')', '').split(',')
            x1, y1, x2, y2 = map(float, coords)
            edges.append(((x1, y1), (x2, y2)))
    return cost, v, e, edges

def get_polygon_cycle(edges):
    """
    Given a list of edges (each edge is a tuple of two vertices),
    returns an ordered list of vertices representing the cycle.
    Assumes that every vertex has exactly two neighbors.
    """
    graph = defaultdict(list)
    for v1, v2 in edges:
        graph[v1].append(v2)
        graph[v2].append(v1)

    # Start from an arbitrary vertex.
    start = next(iter(graph))
    cycle = [start]
    prev = None
    current = start

    while True:
        # Choose the neighbor that is not the vertex we just came from.
        next_vertex = None
        for neigh in graph[current]:
            if neigh != prev:
                next_vertex = neigh
                break
        if next_vertex is None:
            break
        cycle.append(next_vertex)
        if next_vertex == start:
            break
        prev, current = current, next_vertex

    return cycle

def is_valid_polygon(edges):
    """
    Checks whether the provided edges form a valid polygon.
    Returns a tuple: (True/False, message).
    """
    # Build a graph: count the occurrences of each vertex.
    graph = defaultdict(list)
    for v1, v2 in edges:
        graph[v1].append(v2)
        graph[v2].append(v1)

    # Every vertex in a simple closed polygon must have exactly two incident edges.
    for vertex, neighbors in graph.items():
        if len(neighbors) != 2:
            return False, f"Vertex {vertex} does not have degree 2."

    # Retrieve the cycle ordering.
    cycle = get_polygon_cycle(edges)
    if len(cycle) < 3:
        return False, "Not enough vertices to form a polygon."

    # Check that the cycle is closed and covers all vertices.
    if cycle[0] != cycle[-1]:
        return False, "Cycle is not closed."
    if len(cycle) != len(graph) + 1:
        return False, "Cycle does not cover all vertices."

    # Optionally, check for self-intersections using Shapely (if installed).
    try:
        from shapely.geometry import Polygon
        poly = Polygon(cycle[:-1])  # Remove duplicate endpoint for Polygon
        if not poly.is_valid:
            return False, "Polygon is self-intersecting or otherwise invalid."
    except ImportError:
        # Shapely is not installed, so we skip the self-intersection test.
        pass

    return True, "Polygon is valid."

def plot_input_output(stellar_crystals, void_mines, edges):
    # Create a Plotly figure.
    fig = go.Figure()

    # --- Plot Stellar Crystals ---
    stellar_x = [x for x, y, c in stellar_crystals]
    stellar_y = [y for x, y, c in stellar_crystals]
    fig.add_trace(go.Scatter(
        x=stellar_x,
        y=stellar_y,
        mode='markers',
        marker=dict(color='green', size=10, opacity=0.7),
        name='Stellar Crystals'
    ))

    # --- Plot Void Mines ---
    void_x = [x for x, y, m in void_mines]
    void_y = [y for x, y, m in void_mines]
    fig.add_trace(go.Scatter(
        x=void_x,
        y=void_y,
        mode='markers',
        marker=dict(color='red', size=10, opacity=0.7, symbol='x'),
        name='Void Mines'
    ))

    # --- Plot Edges ---
    # For Plotly, separate each segment by None so they are not connected.
    edge_x = []
    edge_y = []
    for (x1, y1), (x2, y2) in edges:
        edge_x.extend([x1, x2, None])
        edge_y.extend([y1, y2, None])
    fig.add_trace(go.Scatter(
        x=edge_x,
        y=edge_y,
        mode='lines',
        line=dict(color='black', width=2),
        name='Edges'
    ))

    # --- Check and Plot the Polygon (if valid) ---
    valid, message = is_valid_polygon(edges)
    print("Polygon validity:", message)
    if valid:
        # Get the cycle ordering.
        cycle = get_polygon_cycle(edges)
        # Remove duplicate last vertex (which equals the first).
        if cycle[0] == cycle[-1]:
            cycle = cycle[:-1]
        # Close the loop for plotting (Plotly needs the first vertex again).
        polygon_x = [p[0] for p in cycle] + [cycle[0][0]]
        polygon_y = [p[1] for p in cycle] + [cycle[0][1]]
        fig.add_trace(go.Scatter(
            x=polygon_x,
            y=polygon_y,
            mode='lines',
            fill='toself',
            fillcolor='rgba(0,0,255,0.1)',
            line=dict(color='blue'),
            name='Polygon'
        ))
    else:
        # Optionally, you might want to show a text annotation if the polygon is invalid.
        fig.add_annotation(
            text=message,
            xref="paper", yref="paper",
            x=0.5, y=0.95, showarrow=False,
            font=dict(color="red", size=16)
        )

    # --- Update Layout ---
    fig.update_layout(
        title="Stellar Crystals, Void Mines, Edges, and Polygon Check",
        xaxis_title="X",
        yaxis_title="Y",
        width=1200,
        height=1200
    )

    fig.show()

# --- Main Execution ---
input_file = 'input00.txt'
output_file = 'output00.txt'

# Read input and output data.
stellar_crystals, void_mines = read_input(input_file)
cost, v, e, edges = read_output(output_file)

# Plot the data and check polygon validity.
plot_input_output(stellar_crystals, void_mines, edges)
```
